```python
# distribution calculation from homework 1

# read the text file by providing the path
def read_file(filename):
    with open(filename, 'r') as file:
        text = file.read()
    return text

# construct a dictionary with the characters as keys and the number of probability of each character as values
def construct_dict(text):
    char_dict = {}
    # iterate through the text and add the characters to the dictionary
    for c in text:
        if c in char_dict:
            char_dict[c] += 1
        else:
            char_dict[c] = 1
    # calculate the probability of each character
    for c in char_dict:
        char_dict[c] = char_dict[c] / len(text)
    return char_dict

# calculate the entropy of the text from the dictionary
def calc_entropy(dict):
    # calculate the entropy
    entropy = 0
    for c in dict:
        p = dict[c]
        entropy -= p * math.log(p, 2)
    return entropy
```

```python


def build_huffman_tree(prob_dict):
    # define a node class to store the character and the probability
    class Node:
        def __init__(self, char=None, prob=0):
            self.char = char
            self.prob = prob
            self.left = None
            self.right = None

    # init the nodes
    nodes = [Node(char, prob) for char, prob in prob_dict.items()]

    # combine the nodes according to the probability
    while len(nodes) > 1:
        # sort by probability
        nodes.sort(key=lambda x: x.prob)
        # combine the two nodes with the smallest probability
        new_node = Node(prob=nodes[0].prob + nodes[1].prob)
        # link the sum of node to the previous two nodes
        new_node.left, new_node.right = nodes.pop(0), nodes.pop(0)
        # add the new node to the list
        nodes.append(new_node)

    return nodes[0]


def generate_huffman_codes(root):
    def traverse_node(node, code, codes):
        # transverse until the node is a leaf node
        if node.char:
            codes[node.char] = code
        else:
            traverse_node(node.left, code + '0', codes)  # transverse L
            traverse_node(node.right, code + '1', codes)  # transverse R

    codes = {}
    traverse_node(root, '', codes)
    return codes


```

```python
# text = read_file('/Users/cyf/Downloads/input.txt')
filename = input("Enter the filename: ")
text = read_file(filename)
```

```python
prob_dict = construct_dict(text)
huffman_tree = build_huffman_tree(prob_dict)
codes = generate_huffman_codes(huffman_tree)

print(codes)
```

```python
# write an encoder to compress text files
def encoder(text, codes):
    encoded_text = ''
    for char in text:
        encoded_text += codes[char]
    return encoded_text

# write a decoder to decompress text files
def decoder(encoded_text, codes):
    decoded_text = ''
    while encoded_text:
        for code in codes:
            if encoded_text.startswith(codes[code]):
                # print(code, codes[code])
                decoded_text += code
                encoded_text = encoded_text[len(codes[code]):]
                break
    return decoded_text
```

```python
# Encode
encoded_text = encoder(text, codes)
print(encoded_text)
```

```python
# Decode
decoded_text = decoder(encoded_text, codes)
print(decoded_text)
```
